```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# First argument is mean
# Second argument is standard deviation
# Third argument is size

# mean and standard deviation
x = np.random.normal(50, 5, 3000)  # Draw random samples from a normal (Gaussian) distribution
```

```python
x
```

```python
print("Max is {}".format(max(x)))
print("Min is {}".format(min(x)))
print("Max - Min is {}".format(max(x) - min(x)))
```

# Frequency Histogram  

Histograms are likely familiar, and a hist function already exists in matplotlib. 

```python
sns.distplot(x, kde=False, bins=5,
             hist_kws=dict(edgecolor="K", linewidth=2))
plt.xlabel('Intervals')
plt.ylabel('Frequency')
plt.show()
```

## Relative Frequency Histogram Plot

```python
sns.distplot(x, kde=False, bins=30,
             norm_hist=True,
             hist_kws=dict(edgecolor="k", linewidth=2))
plt.xlabel('Intervals')
plt.ylabel('Relative Frequency')
plt.show()
```

# Density Plot

A density plot is a representation of the distribution of a numeric variable. It uses a kernel density estimate to show the probability density function of the variable. It is a smoothed version of the histogram and is used in the same concept.

```python
sns.distplot(x, kde=True, bins=30,
             norm_hist=True,
             hist_kws=dict(edgecolor="r", linewidth=2))
plt.xticks(range(28, 70, 4))

plt.ylabel('Density')
plt.show()
```

```python
plt.axvline(x.mean(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(np.median(x), color='r', linestyle='dashed', linewidth=4)
sns.distplot(x, kde=True, bins=30, hist=False)
plt.ylabel('Density')
plt.show()
```
